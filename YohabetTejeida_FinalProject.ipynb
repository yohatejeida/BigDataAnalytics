## <font color='pink'> Q1. NEW YORK BABIES </font>

```python
#Import Pandas matplot and Seaborn
import pandas as pd
pd.set_option('display.precision', 2)
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import and Read CSV 
df=pd.read_csv('/Users/yohatejeida/Desktop/births.csv') 
# Print First 5 Values to Confirm data 
df.head()
```

```python
#Clean and Reformat Data 

# Remove spaces in column names
df.columns = df.columns.str.replace(' ', '') 
# Reformat columns to be in lower case 
df.columns = df.columns.str.lower()
# Delete duplicates 
df = df.drop_duplicates()

# Print Last 5 Values to Confirm Changes 
df.tail()
```

```python
# Rename ethnic groups
df['ethnicity'] = df['ethnicity'].replace(['asian and paci','black non hisp','white non hisp'],['asian and pacific islander', 'black non hispanic','white non hispanic'])
df.head()
```

```python
#Check for null data
df.isnull().sum() * 100 / len(df)
```

Note: After cleaning and reformating the dataset, we have found out that there are no missing values on the dataset. 

```python
# Count the total number of babies born in the dataset

df.frequency.sum()
```

Note: In this dataset, there is a total of 416,699 reported born babies in New York. 

```python
# Group and Count the total number of babies born in the dataset by gender 
df.groupby("gender").frequency.sum()
```

Note: In this dataset, there is a total of 182,574 reported born female babies in New York and 234,125 reported born male babies.

```python
# Plot and compare the total number of babies born by gender in the dataset
sns.set_style("dark")
plt.style.use('fivethirtyeight')
plt.ylabel("Total Babies Born")
df.groupby("gender").frequency.sum().plot(figsize = (8,5), kind="bar",
                                 title ="Total Number of Babies Born by Gender", color = ["pink", "purple"]);
```

Note: This graph displays the total number of babies registred by gender. As you can see the is a significat differece between male and female birth rates

## <font color='pink'> Q2. HOUSING PRICES </font>

```python
# Import and Read CSV 
df=pd.read_excel('/Users/yohatejeida/Desktop/house_prices.xlsx') 
# Print First 5 Values to Confirm data 
df.head()
```

```python
#Clean and Reformat Data 

# Remove spaces in column names
df.columns = df.columns.str.replace(' ', '') 
# Reformat columns to be in lower case 
df.columns = df.columns.str.lower()
# Delete duplicates 
df = df.drop_duplicates()

# Print Last 5 Values to Confirm Changes 
df.tail()
```

```python
df.price.describe()
```

```python
# Plot Histogram on house prices 
sns.set_style("dark")
plt.style.use('fivethirtyeight')
sns.boxplot(data=df, x="price",  color = 'pink');
```

As shown on the above box plot we can see that the mean housing price for the area is $389. 

Moreover, we also know that most houses are valued between $315 and $438 thousand dollars. 

Lastly, we are able to see the few outliers, which are the include the most expensive house valued at $749 thousand

```python
#Group and Count the number of houses by number of bathrooms
number_baths = df.groupby(['baths']).baths.sum()
number_baths.head(6)
```

```python
#Plot the data 
sns.set_style("dark")
plt.style.use('fivethirtyeight')
plt.ylabel("Number of Houses")
number_baths.plot(figsize = (15,5), kind="bar",
                                 title ="Number of Baths",color=['pink', 'magenta','purple', 'grey', 'cyan','blue' ]);
```

As shown above there are 7 houses with 1.0  bath, 2 with 1.5 baths, 24 houses with 2 baths, 15 houses with 2.5 baths, 9 houses with 3 baths, and 4 houses with 3.5 baths 

```python
# create correlation
pricing_corr = df.corr()
pricing_corr
```

```python
# create heat map sns.heatmap
sns.heatmap(data=pricing_corr, annot=True, cmap ="rainbow").set_title("Correlation Matrix");
```

As you can see from the correlation above, there attributes that have the most influence on the pricing of the house are: the square feet, followed by either the number of acres or whether or not the house has a garage. 

We can also note that the attributes that have the least impact on the prices are the age of the house, and surprisingly, the number of rooms in the hosue.

```python
# Analyzing the relationship between price and living space. 
df.plot(x="size_sqft", y="price", kind="scatter", figsize = (10,5), color="purple",
         title="Price vs. Living Space ");
```

As you can see from the plot above, there is a positive correlation between the living space of the house and the house price. Which is also not surprising at all because that is something that most people look for when buying a house. 

It is worth noticing that there are a couple of outliers that could prove that the size of the house shouldn't play a big contribution to the descicion making when the house itself is your perfect home. 

```python
# Analyzing the relationship between price and property size. 
df.plot(x="acres", y="price", kind="scatter", figsize = (10,5), color="blue",
         title="Price vs. Property Size");
```

As you can see from the plot above, there is a positive correlation between the property size and the house price. Which is also not surprising at all because that is something that most people look for when buying a house. Also, the larger your property the more expensive it should be. 

It is worth noticing that there are a couple of outliers that show that sometimes the size of the house shouldn't play a big contribution to the descicion making when the house itself is big enough for you/your family. 

```python
# Analyzing the relationship between price and age. 
df.plot(x="age", y="price", kind="scatter", figsize = (10,5), color="Magenta",
         title="Price vs.  Age");
```

As you can see from the plot above, there is a negative correlation between the age of the house and the house price. Which is somewhat surprising as older building tend to have more problems in the long run, which I think it's worth considering when selling a house. Nevertheless, it could also mean that the houses might be old, but they can be renovated from the inside which would make sense for this scenario.

With a .83 correlation and based on these scenarios I belive that the house's square feet has the most influence when it comes to set up a price for the house. 

```python
# Analyzing the relationship between the number of bathrooms and living space. 
df.plot(x="size_sqft", y="baths", kind="scatter", figsize = (10,5), color="red",
         title="Nbr of Bathrooms vs. Living Space ");
```
